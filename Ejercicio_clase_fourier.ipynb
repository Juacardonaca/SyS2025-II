```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
B = 2   # altura máxima
d1 = 2  # semiancho
T = 2*d1  # ancho total

# Definición con tus ecuaciones
def x_t(t):
    if -d1 <= t <= 0:
        return (B/d1) * t + B         # m1
    elif 0 <= t <= d1:
        return -(B/d1) * t + B        # m2
    else:
        return 0

# Vector de tiempo
t = np.linspace(-T, T, 1000)
x = np.array([x_t(val) for val in t])

# Gráfica
plt.figure(figsize=(8,4))
plt.plot(t, x, 'r', label='$x(t)$')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title("Señal triangular")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import sympy as sp

# Definir símbolos
t, B, d1 = sp.symbols('t B d1', real=True)

# Pendientes
m1 = B/d1     # pendiente positiva en el tramo izquierdo
m2 = -B/d1    # pendiente negativa en el tramo derecho

# Rectas
x_left  = m1*t + B    # tramo -d1 <= t <= 0
x_right = m2*t + B    # tramo 0 <= t <= d1

# Función por tramos
x_t = sp.Piecewise(
    (x_left, (t >= -d1) & (t <= 0)),
    (x_right, (t > 0) & (t <= d1)),
    (0, True)
)

sp.pretty_print(x_t)
```
